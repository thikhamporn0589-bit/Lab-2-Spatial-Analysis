พื้นที่ศึกษาที่เลือกคือ **อำเภอเมืองลำปาง จังหวัดลำปาง** ซึ่งตั้งอยู่ในภาคเหนือของประเทศไทย และเป็นพื้นที่ที่มีลักษณะภูมิประเทศและการใช้ประโยชน์ที่ดินที่หลากหลาย ทั้งพื้นที่ป่าไม้ พื้นที่เกษตรกรรม และเขตเมือง ทำให้สามารถสะท้อนความแตกต่างเชิงพื้นที่ (spatial variation) ได้อย่างชัดเจน  

ลำปางยังเป็นพื้นที่ที่มักเผชิญกับปัญหาภัยแล้งในช่วงฤดูแล้งของทุกปี ส่งผลกระทบต่อทั้งทรัพยากรน้ำและพืชพรรณในพื้นที่ การเลือกพื้นที่นี้จึงมีความน่าสนใจในเชิงภูมิศาสตร์ เนื่องจากสามารถศึกษาความสัมพันธ์ระหว่าง “ความชื้น” และ “ความสมบูรณ์ของพืชพรรณ” ได้อย่างชัดเจน  

นอกจากนี้ ผู้ศึกษาเคยมีประสบการณ์เดินป่าและขี่รถสำรวจพื้นที่ในจังหวัดลำปาง ทำให้ได้สัมผัสลักษณะภูมิประเทศจริง ทั้งพื้นที่ภูเขาและพื้นที่เกษตร ซึ่งเป็นความทรงจำที่ดี และช่วยให้เกิดความเข้าใจเชิงพื้นที่มากขึ้น การเลือกพื้นที่นี้จึงไม่ได้มีเพียงเหตุผลทางวิชาการเท่านั้น แต่ยังมีมิติของประสบการณ์ส่วนตัวที่ช่วยสนับสนุนการตีความข้อมูลได้ลึกยิ่งขึ้น

## Satellite and Temporal Selection

###  Temporal Selection

เลือกช่วงเวลาในการวิเคราะห์เป็น 2 ช่วง ได้แก่  

- **ต้นฤดูแล้ง (พฤศจิกายน – ธันวาคม 2023)**  
- **ปลายฤดูแล้ง (มีนาคม – เมษายน 2024)**  

เหตุผลในการเลือกช่วงเวลาดังกล่าว คือเพื่อเปรียบเทียบระหว่างช่วงที่พื้นที่ยังมีความชื้นสะสมจากฤดูฝน กับช่วงที่สภาพแห้งแล้งรุนแรงที่สุด ซึ่งจะช่วยให้เห็นการเปลี่ยนแปลงของพืชพรรณและความชื้นได้อย่างชัดเจน  

ช่วงเวลาอื่น เช่น ฤดูฝน ไม่ถูกเลือก เนื่องจากมีเมฆปกคลุมจำนวนมาก ส่งผลให้คุณภาพของภาพดาวเทียมลดลง และอาจรบกวนการวิเคราะห์ข้อมูล

### Cloud Filtering

กำหนดเกณฑ์การกรองเมฆที่ **Cloud Cover < 10%**  
เพื่อให้ได้ภาพที่มีความชัดเจนและลดผลกระทบจากเมฆที่บดบังพื้นผิวโลก  

อย่างไรก็ตาม การตั้งเกณฑ์ที่เข้มงวดนี้ทำให้จำนวนภาพที่สามารถนำมาใช้ได้ลดลง ซึ่งเป็น trade-off ระหว่าง “คุณภาพของภาพ” และ “ปริมาณข้อมูล” โดยการเลือกเกณฑ์นี้เน้นให้ได้ข้อมูลที่มีความน่าเชื่อถือสูงสำหรับการวิเคราะห์

```python
import ee
import geemap

# =========================
# 🔐 Initialize
# =========================
ee.Authenticate()
ee.Initialize(project='ee-thikhamporn0589')

# =========================
# 📍 ROI + Districts
# =========================
thailand = ee.FeatureCollection("FAO/GAUL/2015/level2")

roi = thailand.filter(ee.Filter.eq('ADM2_NAME', 'Muang Lampang'))
roi = ee.FeatureCollection([roi.first()])

districts = thailand.filter(ee.Filter.eq('ADM1_NAME', 'Lampang'))

# =========================
# 🛰️ Scale Function
# =========================
def scale(image):
    optical = image.select('SR_B.').multiply(0.0000275).add(-0.2)
    return image.addBands(optical, None, True)

# =========================
# 🌱 Early / 🔥 Late Composite
# =========================
def get_composite(start, end):
    return (ee.ImageCollection("LANDSAT/LC08/C02/T1_L2")
            .filterBounds(roi)
            .filterDate(start, end)
            .filter(ee.Filter.lt('CLOUD_COVER', 10))
            .map(scale)
            .median()
            .clip(roi))

early = get_composite('2023-11-01', '2023-12-31')
late  = get_composite('2024-03-01', '2024-04-30')

# =========================
# 🌿 Spectral Indices
# =========================
ndvi_early = early.normalizedDifference(['SR_B5','SR_B4']).rename('NDVI_early')
ndvi_late  = late.normalizedDifference(['SR_B5','SR_B4']).rename('NDVI_late')

ndwi_early = early.normalizedDifference(['SR_B3','SR_B5']).rename('NDWI_early')
ndwi_late  = late.normalizedDifference(['SR_B3','SR_B5']).rename('NDWI_late')

ndmi_early = early.normalizedDifference(['SR_B5','SR_B6']).rename('NDMI_early')
ndmi_late  = late.normalizedDifference(['SR_B5','SR_B6']).rename('NDMI_late')

# =========================
# 🔥 Change Detection
# =========================
ndvi_change = ndvi_late.subtract(ndvi_early)
ndwi_change = ndwi_late.subtract(ndwi_early)
ndmi_change = ndmi_late.subtract(ndmi_early)

# =========================
# 🗺️ Visualization
# =========================
Map = geemap.Map()
Map.centerObject(roi, 10)

vis_ndvi = {'min': -1, 'max': 1, 'palette': ['blue','white','green']}
vis_ndwi = {'min': -1, 'max': 1, 'palette': ['brown','white','blue']}
vis_change = {'min': -0.5, 'max': 0.5, 'palette': ['red','white','green']}

# NDVI
Map.addLayer(ndvi_early, vis_ndvi, 'NDVI Early')
Map.addLayer(ndvi_late, vis_ndvi, 'NDVI Late')
Map.addLayer(ndvi_change, vis_change, 'NDVI Change')

# NDWI
Map.addLayer(ndwi_early, vis_ndwi, 'NDWI Early')
Map.addLayer(ndwi_late, vis_ndwi, 'NDWI Late')

# NDMI
Map.addLayer(ndmi_early, vis_ndwi, 'NDMI Early')
Map.addLayer(ndmi_late, vis_ndwi, 'NDMI Late')

# Boundary
Map.addLayer(roi, {'color':'red'}, 'Boundary')

Map
```

▸ คำนวณและแสดงผล Spectral Index ที่เหมาะสมกับพื้นที่ศึกษา (ไม่จำกัดเฉพาะ NDVI)



### ▸ คำนวณและแสดงผล Spectral Index ที่เหมาะสมกับพื้นที่ศึกษา (ไม่จำกัดเฉพาะ NDVI)

ในการศึกษานี้ได้เลือกใช้ดัชนีเชิงสเปกตรัมที่สำคัญ ได้แก่  
- **NDVI (Normalized Difference Vegetation Index)** เพื่อวัดความเขียวของพืชพรรณ  
- **NDWI (Normalized Difference Water Index)** เพื่อบ่งชี้ความชื้นและปริมาณน้ำบนพื้นผิว  
- **NDMI (Normalized Difference Moisture Index)** เพื่อวัดความชื้นภายในพืช  

ผลการวิเคราะห์พบว่า ค่า NDVI, NDWI และ NDMI มีแนวโน้มลดลงในช่วงปลายฤดูแล้ง แสดงให้เห็นว่าทั้งความเขียวของพืชและความชื้นในพื้นที่ลดลงไปในทิศทางเดียวกัน ซึ่งสะท้อนถึงผลกระทบของสภาพแห้งแล้งที่มีต่อระบบนิเวศในพื้นที่ศึกษา  

---

### ▸ เปรียบเทียบ Composite 2 ช่วงเวลาและอธิบายการเปลี่ยนแปลงที่พบ

จากการเปรียบเทียบภาพ Composite ระหว่างช่วงต้นฤดูแล้ง (พฤศจิกายน–ธันวาคม) และปลายฤดูแล้ง (มีนาคม–เมษายน) พบว่า ค่าเฉลี่ย NDVI ลดลงจาก **0.696 เป็น 0.506** แสดงถึงการลดลงของความสมบูรณ์ของพืชพรรณอย่างชัดเจน  

แผนที่ NDVI Change แสดงให้เห็นว่าหลายพื้นที่มีค่าลดลง (แสดงเป็นสีแดง) ซึ่งบ่งชี้ว่าพืชพรรณได้รับผลกระทบจากการขาดน้ำในช่วงปลายฤดูแล้ง การเปลี่ยนแปลงนี้สะท้อนถึงอิทธิพลของฤดูกาล โดยความชื้นที่ลดลงส่งผลให้พืชเกิดความเครียดและลดความเขียวลง

### ▸ สร้าง Zonal Statistics สำหรับหน่วยบริหาร (อำเภอ/ตำบล) ของพื้นที่

ได้ทำการวิเคราะห์ Zonal Statistics ในระดับอำเภอ โดยคำนวณค่าเฉลี่ยของ NDVI และ NDWI ในแต่ละพื้นที่ เพื่อเปรียบเทียบความแตกต่างเชิงพื้นที่  

ผลการวิเคราะห์พบว่า ค่า NDVI และ NDWI มีความแตกต่างกันในแต่ละอำเภอ ซึ่งสะท้อนถึงลักษณะการใช้ที่ดินและสภาพแวดล้อมที่แตกต่างกัน เช่น พื้นที่ที่มีค่า NDVI สูงมักเป็นพื้นที่ป่าไม้หรือเกษตรกรรมที่ยังมีความอุดมสมบูรณ์ ขณะที่พื้นที่ที่มีค่า NDVI ต่ำมักเป็นพื้นที่เมืองหรือพื้นที่ที่มีความเสี่ยงต่อความแห้งแล้ง  

นอกจากนี้ หลายพื้นที่มีค่าดัชนีลดลงในช่วงปลายฤดูแล้ง ซึ่งช่วยระบุพื้นที่ที่ได้รับผลกระทบจากความแห้งแล้งได้อย่างชัดเจนในเชิงพื้นที่

```python
# NDVI mean
mean_early = ndvi_early.reduceRegion(ee.Reducer.mean(), roi, 30)
mean_late  = ndvi_late.reduceRegion(ee.Reducer.mean(), roi, 30)

print("NDVI Early:", mean_early.getInfo())
print("NDVI Late:", mean_late.getInfo())

# NDMI mean
mean_ndmi_early = ndmi_early.reduceRegion(ee.Reducer.mean(), roi, 30)
mean_ndmi_late  = ndmi_late.reduceRegion(ee.Reducer.mean(), roi, 30)

print("NDMI Early:", mean_ndmi_early.getInfo())
print("NDMI Late:", mean_ndmi_late.getInfo())
```

## 📊 Mean Value Analysis

จากการวิเคราะห์ค่าเฉลี่ยของ NDVI พบว่าในช่วงต้นฤดูแล้งมีค่าอยู่ที่ **0.696** ซึ่งอยู่ในระดับสูง สะท้อนให้เห็นว่าพื้นที่ยังมีพืชพรรณที่หนาแน่นและมีความอุดมสมบูรณ์ โดยได้รับอิทธิพลจากความชื้นที่ยังคงหลงเหลือจากฤดูฝนก่อนหน้า  

อย่างไรก็ตาม เมื่อเข้าสู่ช่วงปลายฤดูแล้ง ค่า NDVI ลดลงเหลือ **0.506** โดยลดลงประมาณ **-0.19** แสดงให้เห็นถึงการลดลงของความเขียวของพืชพรรณอย่างชัดเจน ซึ่งสัมพันธ์กับสภาพอากาศที่แห้งแล้งและปริมาณน้ำที่ลดลง แม้ว่าค่า NDVI จะยังคงอยู่ในระดับปานกลางถึงสูง แต่แนวโน้มที่ลดลงนี้สะท้อนถึง “ความเครียดของพืช” ที่เริ่มเกิดขึ้นในพื้นที่ศึกษา  

ในส่วนของ NDMI ซึ่งใช้บ่งชี้ความชื้นในพืช พบว่าค่ามีแนวโน้มลดลงในช่วงปลายฤดูแล้งเช่นเดียวกัน แสดงให้เห็นว่าพืชสูญเสียความชื้นอย่างต่อเนื่อง เมื่อพิจารณาร่วมกับ NDVI จะเห็นได้ว่าทั้ง “ความเขียวของพืช” และ “ความชื้นในพืช” ลดลงไปในทิศทางเดียวกัน  

ดังนั้นจึงสามารถตีความได้ว่า “ความขาดแคลนน้ำ” เป็นปัจจัยสำคัญที่ส่งผลต่อสภาพพืชพรรณในพื้นที่ โดยผลลัพธ์นี้ยังสอดคล้องกับการวิเคราะห์จากแผนที่และข้อมูลเชิงพื้นที่ ที่แสดงให้เห็นแนวโน้มการลดลงของดัชนีในหลายบริเวณของพื้นที่ศึกษา  

โดยสรุป แม้ว่าพืชพรรณจะยังไม่เสื่อมโทรมอย่างรุนแรง แต่แนวโน้มที่ลดลงของทั้ง NDVI และ NDMI บ่งชี้ถึงความเสี่ยงที่เพิ่มขึ้นของผลกระทบจากภาวะแห้งแล้งในอนาคต

```python
combined = ndvi_early.addBands(ndvi_late)\
                     .addBands(ndwi_early)\
                     .addBands(ndwi_late)

zonal_stats = combined.reduceRegions(
    collection=districts,
    reducer=ee.Reducer.mean(),
    scale=30
)

print(zonal_stats.limit(5).getInfo())
```

## 📊 Zonal Statistics Analysis

เพื่อทำความเข้าใจความแตกต่างเชิงพื้นที่ของพืชพรรณและความชื้น ได้มีการคำนวณ Zonal Statistics ในระดับอำเภอ โดยนำค่าเฉลี่ยของ NDVI และ NDWI จากทั้งช่วงต้นฤดูแล้งและปลายฤดูแล้งมาสรุปในแต่ละหน่วยบริหาร  

ผลการวิเคราะห์พบว่า ค่า NDVI และ NDWI มีความแตกต่างกันในแต่ละอำเภอ ซึ่งสะท้อนถึงลักษณะการใช้ที่ดินและสภาพแวดล้อมที่แตกต่างกัน เช่น พื้นที่ที่มีค่า NDVI สูงมักเป็นพื้นที่ป่าไม้หรือพื้นที่เกษตรที่ยังมีความอุดมสมบูรณ์ ขณะที่พื้นที่ที่มีค่า NDVI ต่ำมักเป็นเขตเมืองหรือพื้นที่ที่ได้รับผลกระทบจากความแห้งแล้งมากกว่า  

นอกจากนี้ เมื่อเปรียบเทียบระหว่างสองช่วงเวลา พบว่าหลายพื้นที่มีค่า NDVI และ NDWI ลดลงในช่วงปลายฤดูแล้ง แสดงให้เห็นถึงการลดลงของทั้งความเขียวของพืชและความชื้นในพื้นที่ ซึ่งสอดคล้องกับผลการวิเคราะห์ในระดับภาพรวม (mean value analysis)  

การวิเคราะห์ในระดับอำเภอช่วยให้สามารถระบุพื้นที่ที่มีความเสี่ยงต่อภาวะแห้งแล้งได้ชัดเจนมากขึ้น และแสดงให้เห็นว่าผลกระทบของฤดูแล้งไม่ได้เกิดขึ้นอย่างเท่ากันในทุกพื้นที่ แต่มีความแตกต่างตามลักษณะภูมิประเทศและการใช้ประโยชน์ที่ดิน

```python
combined_sample = ndvi_late.addBands(ndwi_late)

sample = combined_sample.sample(
    region=roi,
    scale=30,
    numPixels=5000
)

correlation = sample.reduceColumns(
    ee.Reducer.pearsonsCorrelation(),
    ['NDVI_late','NDWI_late']
)

print("Correlation:", correlation.getInfo())
```

## 📈 Correlation Analysis

จากการวิเคราะห์ความสัมพันธ์ระหว่าง NDVI และ NDWI ในช่วงปลายฤดูแล้ง พบว่ามีค่าสหสัมพันธ์เท่ากับ **-0.926** ซึ่งเป็นความสัมพันธ์เชิงลบในระดับสูงมาก (strong negative correlation)

ผลลัพธ์นี้หมายความว่า พื้นที่ที่มีค่า NDWI สูง (มีน้ำหรือความชื้นผิวดินสูง) กลับมีค่า NDVI ต่ำ ในขณะที่พื้นที่ที่มีค่า NDWI ต่ำ (แห้ง) กลับมีค่า NDVI สูงกว่า

ปรากฏการณ์นี้สามารถอธิบายได้ว่า พื้นที่ที่มีค่า NDWI สูงมักเป็นแหล่งน้ำ เช่น แม่น้ำ อ่างเก็บน้ำ หรือพื้นที่ชุ่มน้ำ ซึ่งโดยธรรมชาติแล้วจะไม่มีพืชพรรณหนาแน่น จึงทำให้ค่า NDVI ต่ำ

ในทางกลับกัน พื้นที่ที่มีพืชพรรณหนาแน่น เช่น ป่าไม้หรือพื้นที่เกษตร มักมีค่า NDWI ต่ำ เนื่องจาก NDWI ไม่ได้สะท้อนความชื้นในพืชโดยตรง แต่เน้นไปที่น้ำผิวดิน

ดังนั้น ความสัมพันธ์เชิงลบที่พบไม่ได้ขัดแย้งกับหลักการทางภูมิศาสตร์ แต่สะท้อนถึงความแตกต่างของลักษณะพื้นผิว (land cover) ระหว่าง “แหล่งน้ำ” และ “พื้นที่พืชพรรณ”

ผลลัพธ์นี้ช่วยยืนยันว่าการใช้ NDWI เพียงอย่างเดียวอาจไม่เพียงพอในการอธิบายความชื้นของพืช และจำเป็นต้องพิจารณาร่วมกับดัชนีอื่น เช่น NDVI และ NDMI เพื่อให้เข้าใจสภาพพื้นที่ได้อย่างครบถ้วน

```python
# Image
task_img = ee.batch.Export.image.toDrive(
    image=ndvi_late,
    description='NDVI_Late',
    folder='GEE',
    region=roi.geometry(),
    scale=30,
    crs='EPSG:4326',
    maxPixels=1e13
)
task_img.start()

# Table
task_csv = ee.batch.Export.table.toDrive(
    collection=zonal_stats,
    description='Zonal',
    folder='GEE',
    fileFormat='CSV'
)
task_csv.start()
```

##  Data Export

ในการศึกษานี้ ได้ทำการส่งออกข้อมูลทั้งในรูปแบบภาพ (raster) และข้อมูลเชิงตาราง (vector) ไปยัง Google Drive เพื่อใช้ในการวิเคราะห์และแสดงผลเพิ่มเติม  

### Coordinate Reference System (CRS)

เลือกใช้ระบบพิกัด **EPSG:4326 (WGS84)**  
เนื่องจากเป็นระบบพิกัดมาตรฐานสากลที่ใช้อย่างแพร่หลาย และสามารถนำไปใช้งานร่วมกับซอฟต์แวร์ GIS ต่าง ๆ เช่น QGIS หรือ ArcGIS ได้สะดวก โดยไม่ต้องแปลงพิกัดเพิ่มเติม  

---

### Scale (Spatial Resolution)

กำหนดค่า **scale = 30 เมตร**  
ซึ่งสอดคล้องกับความละเอียดเชิงพื้นที่ของข้อมูล Landsat 8

การเลือกความละเอียดนี้ช่วยให้สามารถรักษารายละเอียดของข้อมูลต้นฉบับได้อย่างเหมาะสม  
อย่างไรก็ตาม มี trade-off คือ:
- ความละเอียดสูง (30 m) → รายละเอียดมาก แต่ไฟล์มีขนาดใหญ่และใช้เวลาในการประมวลผลมากขึ้น  
- หากใช้ความละเอียดต่ำลง → จะลดขนาดไฟล์และเวลาในการประมวลผล แต่สูญเสียรายละเอียดเชิงพื้นที่  

---

###  File Format

ข้อมูลถูกส่งออกใน 2 รูปแบบ ได้แก่:

- **GeoTIFF (.tif)** → สำหรับข้อมูล NDVI (raster)  
  ใช้เก็บข้อมูลเชิงพื้นที่พร้อมพิกัด ทำให้สามารถนำไปเปิดและวิเคราะห์ต่อในโปรแกรม GIS ได้ทันที  

- **CSV (.csv)** → สำหรับข้อมูล Zonal Statistics  
  ใช้เก็บข้อมูลเชิงตาราง เช่น ค่าเฉลี่ย NDVI และ NDWI ในแต่ละอำเภอ เพื่อให้สามารถนำไปวิเคราะห์เชิงสถิติหรือสร้างกราฟเพิ่มเติมได้  

---

### Interpretation
